# 401. Binary Watch

## 思路
注意到，一个灯对应binary上一个数位，亮一个，对应一个1

```python
class Solution(object):
    def readBinaryWatch(self, num):
        """
        :type num: int
        :rtype: List[str]
        """
        return ['%d:%02d' % (hour, minute) for hour in range(12) for minute in range(60) if (bin(hour) + bin(minute)).count('1') == num]
                #if bin(hour).count("1") + bin(minute).count("1") == num]
```

```python
s = Solution()
s.readBinaryWatch(5)
```
